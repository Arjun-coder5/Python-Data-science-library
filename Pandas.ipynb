# ***Pandas Start***

**import pandas**

```python
import pandas as pd
import numpy as np
```

# Series from List :

```python
# String
country = ['India','pak','USA','Nepal','ShriLanka']
pd.Series(country)
```

```python
# Integer :
runs = [13,12,24,56,78,100]
pd.Series(runs)
```

```python
# Custom index :
marks = [45,34,23,45]
subject = ['maths','science','english','social']
pd.Series(marks,index= subject)
```

```python
# Setting a Name :
marks  =pd.Series(marks,index= subject , name  = "Marks")
marks
```

# ***Series from Dict :***

```python
marks = {
    'maths' : 83,
    'english' :93,
    'Science' : 92
}

marks_series = pd.Series(marks,name = 'Marks')
marks_series
```

# ***Series Attributes :***

```python
# Size :
marks_series.size
```

```python
# dtype :
marks_series.dtype
```

```python
# name :
marks_series.name
```

```python
# is_unique :
# marks_series.is_unique
pd.Series([1,1,1,1]).is_unique
```

```python
# index :
marks_series.index
```

```python
# value :
marks_series.values
```
